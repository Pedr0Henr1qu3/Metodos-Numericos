**1.** Faça o gráfico da função

$$\quad f(x, y)=\left(x^2+3 y^2\right) e^{-x^2-y^2}$$

Em seguida represente a função com curvas de nível.



```python
import numpy as np
import matplotlib.pyplot as plt
def f(x, y):
    return (x**2 + 3*y**2) * np.exp(-x**2 - y**2)


x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)


fig = plt.figure(figsize=(12, 6))


ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('f(x, y)')
ax1.set_title('Gráfico 3D da função f(x, y)')

ax2 = fig.add_subplot(122)
contour = ax2.contour(X, Y, Z, levels=20, cmap='viridis')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Curvas de Nível da função f(x, y)')
plt.colorbar(contour, ax=ax2, label='f(x, y)')

plt.tight_layout()
plt.show()

```

**2.** Plote os campos vetoriais:

(a) $\mathbf{F}(x, y)=\left\langle\ln \left(1+y^2\right), \ln \left(1+x^2\right)\right\rangle$

(b) $\mathbf{F}(x, y, z)=\frac{y}{z} \mathbf{i}-\frac{x}{z} \mathbf{j}+\frac{z}{4} \mathbf{k}$

```python
import numpy as np
import matplotlib.pyplot as plt
def F(x, y):
    return np.array([np.log(1 + y**2), np.log(1 + x**2)])

x = np.linspace(-3, 3, 20)
y = np.linspace(-3, 3, 20)
X, Y = np.meshgrid(x, y)
U, V = F(X, Y)

plt.figure(figsize=(8, 6))
plt.quiver(X, Y, U, V, color='b', angles='xy', scale_units='xy', scale=5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Campo Vetorial F(x, y) = (ln(1+y^2), ln(1+x^2))')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.grid(True)
plt.show()
```

```python
import sympy as sp
sp.init_printing()

#Define os simbolos e funções
x,y = sp.symbols('x, y', real=True)
z = sp.Function('z')
z =  y**2 - 2*y*sp.cos(x)

# Deriva
dzdx = z.diff(x)
dzdy = z.diff(y)

#cria as equações
eq1 = sp.Eq(dzdx,0)
eq2 = sp.Eq(dzdy,0)
pts = sp.solve([eq1,eq2])
pts
```

**4.** Determine o volume do sólido abaixo da superficie
$$z=1+x^2 y^2$$
e acima da região limitada por $x=y^2$ e $x=4$. Faça o gráfico de $z$  e da região de integração

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import sympy as sp

# definindo a superficie
z_func = lambda x, y: 1 + x**2 * y**2

# definindo a regiao de integração
y = np.linspace(-2, 2, 400)
x1 = y**2
x2 = 4 * np.ones_like(y)

# grid para criar a superfice
X = np.linspace(-1, 4, 400)
Y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(X, Y)
Z = z_func(X, Y)

# plotagem da superfice e região de integração
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='cool', alpha=0.7)
ax.plot(y**2, y, z_func(y**2, y), color='r', lw=3)
ax.plot(4 * np.ones_like(y), y, z_func(4 * np.ones_like(y), y), color='r', lw=3)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title(r' $z = 1 + x^2 y^2$ e região de integração')
plt.show()

```

**5.** Faça o gráfico e encontre uma aproximação para o comprimento da curva $C$ com equações paramétricas

$$x=e^{-t} \cos 4 t$$
$$y=e^{-t} \operatorname{sen} 4 t$$
$$z=e^{-t}$$

com $0 \leq t \leq 2 \pi$.

Em seguida, encontre o valor exato da integral $\int_C x^3 y^2 z d s$, .

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Definir as funções paramétricas x(t), y(t), z(t)
def x(t):
    return np.exp(-t) * np.cos(4*t)

def y(t):
    return np.exp(-t) * np.sin(4*t)

def z(t):
    return np.exp(-t)

# Definir o intervalo de t
t = np.linspace(0, 2*np.pi, 100)

# Calcular as coordenadas da curva paramétrica
X = x(t)
Y = y(t)
Z = z(t)

# Plotar o gráfico da curva paramétrica
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(X, Y, Z, label='Curva C', color='b')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Gráfico da Curva Paramétrica C')
ax.legend()

# Calcular o comprimento da curva paramétrica usando a regra do trapézio
def integrand(t):
    return np.sqrt((-np.exp(-t) * np.sin(4*t) * 4*np.exp(-t))**2
                   + (np.exp(-t) * np.cos(4*t) * 4*np.exp(-t))**2
                   + (-np.exp(-t) * np.cos(4*t))**2)

length, _ = quad(integrand, 0, 2*np.pi)
print(f"Aproximação para o comprimento da curva C: {length:.4f}")

plt.show()



```
